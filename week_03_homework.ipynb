```python
#Starter code. Do NOT modify this cell
import pandas as pd
import numpy as np

# Read in dataset
df = pd.read_csv('cars.csv',delimiter=';')

# Convert each column to an np.array
car = df.Car.to_numpy()
mpg = df.MPG.to_numpy()
cylinders = df.Cylinders.to_numpy()
displacement = df.Displacement.to_numpy()
horsepower = df.Horsepower.to_numpy()
weight = df.Weight.to_numpy()
acceleration = df.Acceleration.to_numpy()
model = df.Model.to_numpy()
origin = df.Origin.to_numpy()

# create an array of headers
headers = np.array(['Car', 'MPG', 'Cylinders', 'Displacement', 'Horsepower', 'Weight','Acceleration', 'Model', 'Origin'])
```

5. Show how to:

6. What is the correlation coefficient of `mpg` and `weight`? What is a correlaton coefficient?

```python
# Correlation coefficient tells how one one variable is related to another. How one variable change impacts change in another.
# A positive coefficient indicates that both variable moves in same directions i.e. if one increases the other also increases
# A negative coefficient indicates that both variable moves in opposite directions i.e. if one increases the other decreases

# in this example the coeff is negative. If the weigtht of the car is more mpg value decreases and vice versa.

np.corrcoef(mpg,weight)
```
